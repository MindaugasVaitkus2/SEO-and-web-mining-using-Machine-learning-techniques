```python
from gensim.summarization import summarize
from gensim.summarization import keywords
```

```python
import glob
file_names=glob.glob("stories/*.txt")
file_names
```

```python
raw_documents=[]
for file in file_names:
    try:#we will try to catch block to prevent the code from crashing if it can not rewd any of the text
        with open(file,"r",encoding="utf-8") as f:raw_documents.append(f.read())
    except:
        pass
text=print(raw_documents[1])
```

```python
print(summarize(text,word_count=100))
```
